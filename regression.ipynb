#Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.

- Answer:
Simple Linear Regression (SLR) is a statistical technique used to model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to observed data.
- Purpose:

1.To understand how changes in X affect Y

2.To predict the value of Y for a given X

3.To identify trends and relationships between variables

#Question 2: What are the key assumptions of Simple Linear Regression?

- Answer:
The key assumptions are:

1.Linearity – Relationship between X and Y is linear

2.Independence – Observations are independent of each other

3.Homoscedasticity – Constant variance of errors

4.Normality of errors – Residuals are normally distributed

5.No multicollinearity – Only one independent variable is used

#Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

- Answer:

y=mx+c

Where:

y = dependent variable

x = independent variable

m = slope (change in y per unit change in x)

c = intercept (value of y when x = 0)

#Question 5: What is the method of least squares in linear regression?

- Answer:
The method of least squares finds the best-fitting line by minimizing the sum of the squared differences between actual values and predicted values. It ensures the most accurate line through the data points.

#Question 7: Name and briefly describe three common evaluation metrics for regression models.

- Answer:

1.Mean Squared Error (MSE): Average of squared prediction errors

2.Mean Absolute Error (MAE): Average of absolute errors

3.R-squared (R²): Proportion of variance explained by the model

#Question 8: What is the purpose of the R-squared metric in regression analysis?

- Answer:

R-squared measures how well the independent variable explains the variability of the dependent variable.

- Value ranges from 0 to 1

- Higher value indicates a better model fit

#Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

- Answer:



```python
from sklearn.linear_model import LinearRegression
import numpy as np


X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])


model = LinearRegression()
model.fit(X, y)


print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```


#Question 10: How do you interpret the coefficients in a simple linear regression model?

Answer:

- Slope (m): For every one-unit increase in X, Y increases by m units

- Intercept (c): Expected value of Y when X is zero
